```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
docs = [
    "The Eiffel Tower is a wrought-iron lattice tower in Paris, France.",
    "Python is a high-level programming language known for its readability.",
    "FAISS is a library for efficient similarity search and clustering of dense vectors.",
    "Chroma is an open-source embedding database designed for AI applications.",
    "Sentence Transformers allow easy computation of dense vector representations for text.",
    "The Great Wall of China is a historic fortification built to protect Chinese states.",
    "Machine learning enables systems to learn patterns from data without explicit programming.",
    "Mount Everest is the highest mountain above sea level, located in the Himalayas."
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(docs)
```

```python
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
```

```python
query = "What is the tallest mountain?"
query_embedding = model.encode([query])
```

```python
D, I = index.search(query_embedding, k=3)
print("FAISS Results:")
for idx in I[0]:
    print("-", docs[idx])
```
