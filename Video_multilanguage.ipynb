````python
import moviepy.editor as mp

# Define the input video file path (the public video that was successfully downloaded)
video_file_public = "input_video_public.mp4"
# Define the output audio file path
audio_path_public = "output_audio_public.mp3"

# Extract audio using moviepy
video_clip_public = mp.VideoFileClip(video_file_public)
audio_clip_public = video_clip_public.audio
audio_clip_public.write_audiofile(audio_path_public)

# Close the clips to release resources
audio_clip_public.close()
video_clip_public.close()

print(f"Audio extracted and saved to {audio_path_public}")
````

````python
# Load the Whisper model and transcribe the audio
model = whisper.load_model("small")  # You can choose "base", "medium", or "large" based on your needs
result = model.transcribe(audio_path_public)
transcription = result["text"]
print("Original Transcription:\n", transcription)
````

````python
# Save the transcription to a text file
output_text_file = "transcription.txt"
with open(output_text_file, "w", encoding="utf-8") as f:
    f.write(transcription)
print(f"Transcription saved to {output_text_file}")
````
