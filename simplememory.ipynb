```python
import requests
from bs4 import BeautifulSoup
```

```python
class SimpleAgent:
    def __init__(self):
        # local memory dictionary
        self.memory = {}

    def remember(self, key, value):
        """Store a fact in memory"""
        self.memory[key] = value
        print(f"Remembered: {key} = {value}")

    def recall(self, key):
        """Retrieve a fact"""
        return self.memory.get(key, "Not remembered yet.")

    def fetch_and_summarize(self, url, max_sentences=3):
        """Fetch webpage and summarize"""
        try:
            response = requests.get(url, timeout=10)
            response.raise_for_status()
        except Exception as e:
            return f"Error fetching: {e}"
        soup = BeautifulSoup(response.text, "html.parser")

        paragraphs = [p.get_text() for p in soup.find_all("p")]
        text = " ".join(paragraphs)

        sentences = text.split(".")
        summary = ". ".join(sentences[:max_sentences]) + "..."
        return summary
```

```python
agent = SimpleAgent()
```

```python
agent.remember("favorite_color", "blue")
print("Recall:", agent.recall("favorite_color"))
```

```python
url = "https://share.google/9NQCZ8cuQjOcQQ6sS"
print("\nSummary of AI page:")
print(agent.fetch_and_summarize(url))
```
