```python
import speech_recognition as sr
from google.colab import files
```

```python
print("Please upload an audio file (.wav or .mp3)")
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
print("Uploaded file:", filename)
```

```python
from pydub import AudioSegment

recognizer = sr.Recognizer()

# Check if the file is .m4a and convert if necessary
if filename.endswith('.m4a'):
    print(f"Converting {filename} to WAV format...")
    audio_segment = AudioSegment.from_file(filename, format="m4a")
    wav_filename = filename.replace('.m4a', '.wav')
    audio_segment.export(wav_filename, format="wav")
    file_to_process = wav_filename
    print(f"Conversion complete. Processing {wav_filename}")
else:
    file_to_process = filename

with sr.AudioFile(file_to_process) as source:
    audio = recognizer.record(source)  # read the entire file

    try:
        text = recognizer.recognize_google(audio, language="en-US")  # you can change language here
        print("Recognized Text:", text)
    except sr.UnknownValueError:
        print("Could not understand audio")
    except sr.RequestError:
        print("Error with the recognition service")
```
